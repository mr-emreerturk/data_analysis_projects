__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

2. __Filter__ the Dataset and __find the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

__Define__ an appropriate __user-defined function__ to reuse code.

```python
df_best = df[["poster_path", "title", "budget_musd", "revenue_musd",
              "vote_count", "vote_average", "popularity"]].copy()
df_best.head(2)
```

```python
df_best["profit_musd"] = df.revenue_musd.sub(df.budget_musd)
df_best["return"] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best.columns = ["", "Title", "Budget", "Revenue", "Votes", 
                   "Average Rating", "Popularity", "Profit", "ROI"]
df_best.set_index("Title", inplace=True)
```

```python
df_best.head(2)
```

```python
subset = df_best.iloc[:5, :2]
HTML(subset.to_html(escape=False))
```

```python
df_best.Budget.fillna(0, inplace = True)
df_best.Votes.fillna(0, inplace = True)
```

```python
df_best.info()
```

```python
def best_worst(n, by, ascending = False, min_bud = 0, min_votes = 0):
    
    
    df2 = df_best.loc[(df_best["Budget"] >= min_bud) & (df_best["Votes"] >= min_votes), 
                      ["", by]].sort_values(by = by, ascending = ascending).head(n).copy()
    
    return HTML(df2.to_html(escape=False))
```

__Movies Top 5 - Highest Revenue__

```python
best_worst(n=5, by="Revenue")
```

__Movies Top 5 - Highest Budget__

```python
best_worst(5, "Budget")
```

__Movies Top 5 - Highest Profit__

```python
best_worst(5, "Profit")
```

__Movies Top 5 - Lowest Profit__

```python
best_worst(5, "Profit", ascending=True)
```

__Movies Top 5 - Highest ROI__

```python
best_worst(5, "ROI", min_bud=5)
```

__Movies Top 5 - Lowest ROI__

```python
best_worst(5, "ROI", ascending=True, min_bud=10)
```

__Movies Top 5 - Most Votes__

```python
best_worst(5, "Votes", min_votes=1000)
```

__Movies Top 5 - Highest Rating__

```python
best_worst(5, "Average Rating", min_votes=1000)
```

__Movies Top 5 - Lowest Rating__

```python
best_worst(5, "Average Rating", ascending=True, min_votes= 20)
```

__Movies Top 5 - Most Popular__

```python
best_worst(5,"Popularity", min_votes=100)
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
df.head(1)
```

```python
mask_studio = df["production_companies"].str.contains("Pixar Animation Studios")
mask_year = df["release_date"].between("2010-01-01", "2015-12-31")
pixar = df.loc[mask_studio & mask_year, ["title", "poster_path", "revenue_musd", "release_date"]].sort_values(by = "revenue_musd", ascending=False).set_index("title")
HTML(pixar.to_html(escape=False))
```

__Search 4: Action or Thriller Movie with original language English and minimum Rating of 7.5 (most recent movies first)__

```python
mask_genre = df["genres"].str.contains("Action") | df["genres"].str.contains("Thriller")
mask_language = df["original_language"] == "en"
mask_voting = df["vote_average"] >= 7.5
mask_voters = df["vote_count"] > 10

search_4 = df.loc[mask_genre & mask_language & mask_voting & mask_voters
                  , ["title", "poster_path", "vote_average", "vote_count", "release_date"]].sort_values(by="release_date", ascending=False).set_index("title")
HTML(search_4.to_html(escape=False))
```

4. __Analyze__ the Dataset and __find out whether Franchises (Movies that belong to a collection) are more successful than stand-alone movies__ in terms of:

- mean revenue
- median Return on Investment
- mean budget raised
- mean popularity
- mean rating

hint: use groupby()

```python
df["Franchise"] = df.belongs_to_collection.notna()
df.head(1)
```

```python
df.Franchise.value_counts()
```

__Franchise vs. Stand-alone: Average Revenue__

```python
df.groupby("Franchise")["revenue_musd"].mean()
```

__Franchise vs. Stand-alone: Return on Investment / Profitability (median)__

```python
df["ROI"] = df["revenue_musd"].div(df.budget_musd)
df.groupby("Franchise")["ROI"].median()
```

__Franchise vs. Stand-alone: Average Budget__

```python
df.groupby("Franchise")["budget_musd"].mean()
```

__Franchise vs. Stand-alone: Average Popularity__

```python
df.groupby("Franchise")["vote_average"].mean()
```

__Franchise vs. Stand-alone: Average Rating__

```python
df.groupby("Franchise")["vote_average"].mean()
```

### Summary of all steps in one

```python
df.groupby("Franchise").agg({"budget_musd": "mean", "revenue_musd": "mean", "vote_average": "mean",
                            "popularity": "mean", "ROI":"median", "vote_count":"mean"})
```

## Most Successful Franchises

5. __Find__ the __most successful Franchises__ in terms of

- __total number of movies__
- __total & mean budget__
- __total & mean revenue__
- __mean rating__

```python
franchises = df.groupby("belongs_to_collection").agg({"title":"count", "budget_musd": ["sum", "mean"], 
                                                      "revenue_musd": ["sum", "mean"],
                                                      "vote_average": "mean", "popularity": "mean",
                                                      "ROI":"median", 
                                                      "vote_count":"mean"})

franchises
```

```python
franchises.nlargest(5, ("revenue_musd", "mean"))
```

```python
franchises.nlargest(5, ("title", "count"))
```

```python
franchises.nlargest(5, ("budget_musd", "mean"))
```

```python
franchises[franchises[("vote_count", "mean")] >=1000].nlargest(5, ("vote_average", "mean"))
```
